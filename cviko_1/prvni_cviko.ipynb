```python
# řešení nalezení nejtěžší a nejlehčí kuličky

# Nejdříve si připravíme vstupní seznam - jakékoli váhy kuliček
kulicky = [8,5,2,6,4,1]
```

```python
# funkce range vrací 'generátor' (to ještě nemusíme vědět, co je)
# - pokud ho použijeme ve for cyclu, vrací nám postupně hodnoty od 0 po (horní hranici - 1)
# - v našem případě 0 až (délka pole - 1)
# - délku pole vrací funkce len(kulicky)

for i in range(len(kulicky)):
  print(f"Kulička na indexu {i}: {kulicky[i]}")
```

```python
# Pokud iterujeme od jiné hodnoty než 0, například 2 (3. prvek), dáme funkci dva parametryS

for i in range(2, len(kulicky)):
  print(f"Kulička na indexu {i}: {kulicky[i]}")
```

```python
# Takto můžeme udělat trojice. Pokud by pole nemělo délku dělitelnou 3, museli bychom to ošetřit (to si necháme na příště)

for i in range(len(kulicky) // 3): # i = 0,1...l(k) // 2 -1
  print(3 * i)
  print(3 * i + 1)
  print(3 * i + 2)
```

```python
# poznámka stranou - občas se spletu a řeknu místo 'seznam' slovo 'pole'.

# V jiných programovacích jazycích se používá pole, které je většinou fixní délky,
#   zatímco v pythonu se používá seznam, který se může libovolně zvětšovat pomocí append():

a_list = [1, 2, 3]
print(a_list)

a_list.append(5)
print(a_list)
```

```python
# Rozdělení kuliček na dvě pole - lehčí a těžší

lehci = []
tezsich = []

for i in range(len(kulicky) // 2): # i = 0,1...l(k) // 2 -1
  print(i)
  a, b = kulicky[2 * i], kulicky[2 * i + 1]  # sudý a lichý prvek
  print(f"Comparing element {2 * i} and {2 * i + 1}")

  if a > b:
    lehci.append(b)
    tezsich.append(a)
  else:
    lehci.append(a)
    tezsich.append(b)

lehci, tezsich
```
